Пусть выборка $X_1, ..., X_n$ порождается некоторым распределением $F$. Пусть у распределения $F$ есть некий параметр $\theta$. Например это может быть вероятность успеха в схеме Бернулли, коэффициент эксцесса распределения $F$, или какой-нибудь другой функционал $T(F)$.

Зададимся некоторым числом $\alpha$. Наша цель - написать правило $R(X_1, ..., X_n)$. Это правило будет по выборке выдавать 2 числа - начало и конец промежутка, в котором лежит $\theta$. Этот промежуток называют <b>доверительным интервалом</b>.

Поскольку $R$ зависит от случайных величин, $R$ сама также является случайной величиной. Поэтому возникает ещё одно требование к правилу: для заданного в условии уровня $\alpha$ правило должно выдавать такой промежуток, что $\theta$ лежит в нём с вероятностью $1 - \alpha$. Часто $\alpha = 0.05$, т.е. истинное значение $\theta$ лежит в доверительном интервале с вероятностью $95$%.

Часто интервалы строятся не точно, а приближённо. Вместо каких-то параметров подставляются их оценки по выборке.

0. Пусть $X_1, ..., X_n$ - выборка из распределения Бернулли с параметром $p$. $\hat p = \overline{X}$ - оценка параметра $p$. Постройте $(1 - \alpha)$ доверительные интервалы для оценки p:
    * используя неравенство Чебышёва.
    * асимптотически нормальный доверительный интервал.

Решение: $\Exp \hat p = p$ (честно посчитаем мат. ожидание). Неравенство Чебышёва:
$$
P(\left|\hat p - \Exp \hat p\right| \ge \varepsilon) = P(\left|\hat p - p\right| \ge \varepsilon) \le \frac{\Var \hat p}{\varepsilon^2} = \alpha
$$
$\alpha$ - вероятность того, что истинное значение не попадёт в д.и. Таким образом:
$$
\frac{\sqrt{\Var \hat p}}{\sqrt{\alpha}} = \frac{\se \left(\hat p\right)}{\sqrt{\alpha}} = \varepsilon
$$
Посчитаем дисперсию $\hat p$: $\Var \hat p = \frac{p(1 - p)}{n}$. Поскольку значение $p$ - неизвестно, лучшее, что можно сделать - подставить вместо $p$ его оценку:
$$
\hat{\Var} \hat p = \frac{\hat p(1 - \hat p)}{n}
$$
И получить, что $$\hat \varepsilon = \frac{\sqrt{\frac{\hat p(1 - \hat p)}{n}}}{\sqrt{\alpha}}$$
Подставив это в неравенство Чебышёва, получим:
$$
P(\left|\hat p - p\right| \ge \hat \varepsilon) \le \alpha$$$$
P(\left|\hat p - p\right| < \hat \varepsilon) \ge 1 - \alpha
$$
Тогда доверительный интервал выглядит так:
$$ \left|\hat p - p\right| < \hat \varepsilon$$
$$ - \hat \varepsilon < \hat p - p < \hat \varepsilon$$
$$ \hat p - \hat \varepsilon < p < \hat p + \hat \varepsilon$$
$$ \hat p - \frac{\sqrt{\frac{\hat p(1 - \hat p)}{n}}}{\sqrt{\alpha}} < p < \hat p + \frac{\sqrt{\frac{\hat p(1 - \hat p)}{n}}}{\sqrt{\alpha}}$$
Обратите внимание, д.и. строится именно для $p$, а не для $\hat p$ - мы пытаемся узнать, где находится настоящее значение, а не в какую точку попадёт её оценка по выборке.

Теперь построим асимптотически нормальный доверительный интервал. Рассмотрим следующее выражение:
$$ \frac{\hat p - p}{\se \left(\hat p\right)}$$
Можно показать, что при $n \to +\infty$ эта величина сходится к стандартному нормальному распределению:
$$ \frac{\hat p - p}{\se \left(\hat p\right)} \sim N(0, 1)$$
Для удобства введём $\xi \sim N(0, 1)$. Тогда $\forall x$:
$$P\left(\frac{\hat p - p}{\se \left(\hat p\right)} \le x\right) \approx 
P\left(\xi \le x\right) = \Phi(x)
$$
$$ \Phi(x) = \int\limits_{-\infty}^{x}\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}dx$$
А теперь построим доверительный интервал:
$$ P\left(\left|\frac{\hat p - p}{\se \left(\hat p\right)}\right| \le x\right) \approx 
P\left(\left|\xi\right| \le x\right) = \Phi(x) - \Phi(-x) = 1 - \alpha$$
Т.е. значение лежит в промежутке от $-x$ до $x$ с вероятностью $1 - \alpha$.
Этот $x$ находится так:
$$ x = \Phi^{-1} \left(1 - \frac{\alpha}{2}\right)$$
Теперь можно записать доверительный интервал:
$$\left|\frac{\hat p - p}{\se \left(\hat p\right)}\right| \le \Phi^{-1} \left(1 - \frac{\alpha}{2}\right)$$
Проделав преобразования, подобно предыдущему пункту задачи, получим:
$$\hat p - \se \left(\hat p\right) \cdot \Phi^{-1} \left(1 - \frac{\alpha}{2}\right) \le p \le
\hat p + \se \left(\hat p\right) \cdot \Phi^{-1} \left(1 - \frac{\alpha}{2}\right) 
$$
Поскольку $\se \left(\hat p\right)$ зависит от неизвестного $p$, вместо $\se$ подставим $\estSe$:
$$\hat p - \sqrt{\frac{\hat p(1 - \hat p)}{n}} \cdot \Phi^{-1} \left(1 - \frac{\alpha}{2}\right) \le p \le
\hat p + \sqrt{\frac{\hat p(1 - \hat p)}{n}} \cdot \Phi^{-1} \left(1 - \frac{\alpha}{2}\right) 
$$
Это и есть асимптотически нормальный доверительный интервал.

1.  Дана выборка из экспоненциального распределения с параметром $\lambda$ (Функция распределения: $F(x) = 1 - e^{-\lambda x}$ при $x > 0$, иначе $0$). $\lambda$ считается неизвестным, для генерации данных используем $\lambda = 3$. Используйте seed = 0.

````python
lambda_param = 3
random_state = np.random.RandomState(0)
````

* Сгенерируйте выборку размера $n=1000$ из экспоненциального распределения.

````python
n = 1000
sample = random_state.exponential(1 / lambda_param, n)
````

* Найдите выборочное среднее и выборочную дисперсию.

````python
print(f'Выборочное среднее: {np.mean(sample)}')
print(f'Выборочная дисперсия: {np.std(sample) ** 2}')
````

Постройте графики функции распределения и эмпирической функции распределения при заданном выше $n$. Что можно сказать, глядя на них?

````python
ecdf = ECDF(sample)
cdf = scipy.stats.expon(scale=1 / lambda_param).cdf(ecdf.x)

plt.figure(figsize = (16, 8))
plt.plot(ecdf.x, ecdf.y, label='Эмпирическая функция распределения', color='blue')
plt.plot(ecdf.x, cdf, color='red', label='Истинная функция распределения')
plt.title('Сравнение истинной и эмпирической функции распределения')
plt.xlabel('x')
plt.ylabel('$\hat F(x)$')
plt.grid()
plt.legend()
plt.show()
````

* Постройте $(1-\alpha)$ доверительный интервал для $\lambda$ с помощью оценки $\hat \lambda = \frac{1}{\overline{X}}$.

````python
alpha = 0.05
````

Сперва воспользуемся неравенством Чебышёва:

$$P(\left|\tilde \lambda - \Exp \tilde \lambda\right| \ge \varepsilon) \le \frac{\Var \tilde \lambda}{\varepsilon^2} = \alpha$$

Обратите внимание: здесь тильда, а не крышка. В прошлом примере оценка параметра была несмещённой, поэтому в неравенстве Чебышёва мы просто заменили $\Exp \hat p$ на $p$ (они равны). Здесь оценка $\hat \lambda = \frac{1}{\overline{X}}$, приведённая в условии является смещённой. А именно, можно показать, что
$$ \Exp \hat \lambda = \Exp \frac{1}{\overline{X}} = \frac{n}{n - 1} \lambda$$
Наша цель - найти интервал для $\lambda$, а не для $\hat \lambda$, поэтому можно использовать "исправленную" оценку:
$$ \tilde \lambda = \frac{n - 1}{n} \hat \lambda = \frac{n - 1}{n} \frac{1}{\overline{X}}$$

Коэффициент специально выбран так, чтобы оценка была несмещённой. Чтобы решить задачу, нужно найти дисперсию $\tilde \lambda$. Можно показать, что
$$\Var \tilde \lambda = \frac{\lambda^2}{n - 2}$$

Таким образом, $\varepsilon = \frac{\lambda}{\sqrt{\alpha \lp n - 2\rp}}$ (смотри предыдущий пример). Теперь чтобы построить доверительный интервал, нужно как-то приблизить $\lambda$. Лучше для этого взять какую-нибудь состоятельную оценку. Взяв $\tilde \lambda$ и воспользовавшись рассуждениями из предыдущего примера, получим:
$$ \frac{n - 1}{n} \frac{1}{\overline{X}} - \frac{\frac{n - 1}{n} \frac{1}{\overline{X}}}{\sqrt{\alpha \lp n - 2\rp}} < \lambda < \frac{n - 1}{n} \frac{1}{\overline{X}} + \frac{\frac{n - 1}{n} \frac{1}{\overline{X}}}{\sqrt{\alpha \lp n - 2\rp}}$$

Аналогичным образом получается асимптотически нормальный доверительный интервал:

$$ 
\frac{n - 1}{n} \frac{1}{\overline{X}} - \frac{\frac{n - 1}{n} \frac{1}{\overline{X}}}{\sqrt{n - 2}} \cdot \Phi^{-1} \left(1 -\frac{\alpha}{2}\right) < \lambda < 
\frac{n - 1}{n} \frac{1}{\overline{X}} + \frac{\frac{n - 1}{n} \frac{1}{\overline{X}}}{\sqrt{n - 2}} \cdot \Phi^{-1} \left(1 -\frac{\alpha}{2}\right) 
$$

Теперь напишем код, выдающий доверительные интервалы:

````python
mean = np.mean(sample)
est_lambda = (n - 1) / (n * mean)
est_se = est_lambda / (n - 2) ** 0.5

left_bound = est_lambda - est_se / alpha ** 0.5
right_bound = est_lambda + est_se / alpha ** 0.5
print(f'доверительный интервал на основе неравенства Чебышёва '
      f'для α = {alpha}: ({left_bound}; {right_bound})')

z_alpha = scipy.stats.norm.ppf(1 - alpha / 2, loc=0.0, scale=1.0)
    # 𝛼-квантиль нормального распределения
print(f'𝛼-квантиль нормального распределения: {z_alpha}')
left_bound = est_lambda - z_alpha * est_se
right_bound = est_lambda + z_alpha * est_se
print(f'асимптотически нормальный интервал '
      f'для α = {alpha}: ({left_bound}; {right_bound})')
````

Как можно заметить, истинное значение попадает в доверительный интервал. При этом, интервал на основе неравенства Чебышёва несколько шире.

* Попадает ли истинное значение в доверительный интервал с ростом $n$?

Нужно написать код так, чтобы он работал за $O(n)$ а не $O(n^2)$.

````python
n_start = 3
ns = np.arange(n_start, n + 1, 1)  # размеры префиксов

means = np.cumsum(sample)[n_start - 1:] / ns
real_values = np.tile(np.array([lambda_param, ]), ns.size)
multiplier = (ns - 1) / ns
stds = multiplier / means / (ns - 2) ** 0.5

estimators = multiplier / means

cheb_lower = estimators - stds / alpha ** 0.5
cheb_upper = estimators + stds / alpha ** 0.5

z_alpha = scipy.stats.norm.ppf(1 - alpha / 2, loc=0.0, scale=1.0)
asymptotic_lower = estimators - stds * z_alpha
asymptotic_upper = estimators + stds * z_alpha

plt.figure(figsize = (16, 8))
plt.plot(ns, real_values, color='red', label='Истинное значение')
plt.plot(ns, cheb_upper, color='green')
plt.plot(ns, cheb_lower, color='green', label='Доверительный интервал с помощью неравенства Чебышёва')

plt.fill_between(ns, asymptotic_upper, asymptotic_lower, 
                 color='blue', label='Асимптотически нормальный доверительный интервад')

plt.plot(ns, estimators, color='orange', label='Несмещённая оценка')

plt.title('Доверительные интервалы')
plt.xlabel('$n$')
plt.ylabel('$\lambda$')
plt.grid()
plt.legend()
plt.show()
````

Настоящее значение попадает в доверительный интервал. Оценка сходится к истинному значению (что не противоречит её состоятельности).
